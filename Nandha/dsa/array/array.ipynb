```python
listname = ["apple","banana","cherry","apple","pineapple"]

print("Before changing list",listname)

#length of the list
print("Length of list",len(listname))

#Update the list items
listname[0]=10

print(listname)
#negative index method to print values
print("Last entry of the list is",listname[-1])

#empty list
list1 = []
#getting input from the user with space seperated
#listvalues=input("Enter list items space seperated")

#assigning values from the input with split
#list1=listvalues.split()
#print(list1)

#Append your list
list1.append(1)
list1.append(2)
list1.append(1.5)
list1.append("name")
print("Before changing",list1)

#Insert at specific position
list1.insert(2,"apple")
print("After changing",list1)

#Extend for list
list1.extend([100,200,300,"value1","value2",5.5])
print("After extension of list",list1)

#Reverse
revlist=list(reversed(list1))
print("Reverse list values",revlist)

#remove list items
list1.remove(100)
print("After removing 100",list1)

#removal based on index
list1.pop()
print(list1)
```
