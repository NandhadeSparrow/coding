# Big O Complexity Examples

Big O notation is a mathematical concept used to describe the upper bound of the time complexity or space complexity of an algorithm. It characterizes functions according to their growth rates: how the runtime or space requirements grow relative to the input size. Big O notation provides an upper limit on the growth rate, ensuring that, beyond a certain point, the algorithm will not exceed this limit.

### 1. **O(1) - Constant Time**

```python
# Accessing an element in a list by index is O(1)
def get_first_element(arr):
    return arr[0]  # Always takes the same amount of time
arr = [1, 2, 3, 4, 5]
get_first_element(arr)
```

### 2. **O(log n) - Logarithmic Time**

```python
# Binary search - O(log n)
def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1

arr = [1, 3, 5, 7, 9]
binary_search(arr, 5)
```

### 3. **O(n) - Linear Time**

```python
# Looping through an array - O(n)
def print_elements(arr):
    for element in arr:
        print(element)

arr = [1, 2, 3, 4, 5]
print_elements(arr)
```

### 4. **O(n log n) - Linearithmic Time**

```python
# Merge sort - O(n log n)
def merge_sort(arr):
    if len(arr) <= 1:
        return arr
    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])
    
    return merge(left, right)

def merge(left, right):
    sorted_arr = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            sorted_arr.append(left[i])
            i += 1
        else:
            sorted_arr.append(right[j])
            j += 1
    sorted_arr.extend(left[i:])
    sorted_arr.extend(right[j:])
    return sorted_arr

arr = [5, 3, 8, 4, 2]
merge_sort(arr)
```

```python
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        left_half = arr[:mid]
        right_half = arr[mid:]

        merge_sort(left_half)
        merge_sort(right_half)

        i = j = k = 0

        while i < len(left_half) and j < len(right_half):
            if left_half[i] < right_half[j]:
                arr[k] = left_half[i]
                i += 1
            else:
                arr[k] = right_half[j]
                j += 1
            k += 1

        while i < len(left_half):
            arr[k] = left_half[i]
            i += 1
            k += 1

        while j < len(right_half):
            arr[k] = right_half[j]
            j += 1
            k += 1
```

### 5. **O(n^2) - Quadratic Time**

```python
# Two nested loops - O(n^2)
def print_pairs(arr):
    for i in range(len(arr)):
        for j in range(i + 1, len(arr)):
            print(arr[i], arr[j])

arr = [1, 2, 3, 4]
print_pairs(arr)
```

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
```

```python
L = list()
M = list()
L = set(L) # O(n)
C = []
for i in M:
    if i in L:
        C.append(i)
return C
```

### 6. **O(2^n) - Exponential Time**

```python
# Fibonacci sequence - O(2^n)
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)

fibonacci(5)
```

### 7. **O(n!) - Factorial Time**

```python
# Factorial recursive function - O(n!)
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)

factorial(5)
```
