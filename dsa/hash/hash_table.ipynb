```python
# Hash Table Implementation

class HashTable:
    def __init__(self,size):
        self.size = size
        self.table = [[] for i in range(size)]

    def hash_function(self,key):
        return hash(key) % self.size
    
    def insertion_operation(self,key,value):
        index = self.hash_function(key)

        # check if the key already exists and update it if so
        for kvp in self.table[index]:
            if kvp[0]==key:
                kvp[1] =value
                return
            
        # if the key doesn't exisit, append the new key value pair

        self.table[index].append([key,value])

    def search_operation(self,key):
        index = self.hash_function(key)
        for kvp in self.table[index]:
            if kvp[0]==key:
                return kvp[1]
        return None
    
    def delete_operation(self,key):
        index = self.hash_function(key)
        for i,kvp in enumerate(self.table[index]):
            if kvp[0]==key:
                del self.table[index][i]
                return
            
    def display(self):
        for i,bucket in enumerate(self.table):
            print(f"Bucket{i}:{bucket}")

    def print_all_values(self):
        for bucket in self.table:
            for kvp in bucket:
                print(f"Key:{kvp[0]}, value:{kvp[1]}")


hashtable_obj = HashTable(10)
hashtable_obj.insertion_operation(1,"apple")
hashtable_obj.insertion_operation(2,"banana")
hashtable_obj.insertion_operation(3,"guava")
hashtable_obj.insertion_operation(1,"grape")
hashtable_obj.insertion_operation(4,"watermelon")
hashtable_obj.insertion_operation(5,"muskmelon")
hashtable_obj.insertion_operation(12,"strawberry")
hashtable_obj.insertion_operation(15,"blueberry")

print("Initial state of the hash table")
hashtable_obj.display()

print("\nPrinting all the values in the hash table")
hashtable_obj.print_all_values()

print("\n Search for 2")
print(f"Value:{hashtable_obj.search_operation(2)}")

print("\nDeleting 2")
hashtable_obj.delete_operation(2)

print("\n State of the hash table after deleting 2")
hashtable_obj.display()

print("\nPrint all the values in the hash table after deletion of 2")
hashtable_obj.print_all_values()



            
```
