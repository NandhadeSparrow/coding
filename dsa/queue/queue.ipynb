```python
# Queue Data Structure Implementation

from queue import Queue

# Creation of queue with maxsize of 5
q = Queue(maxsize=5)

print("Queue size before inserting any element: ", q.qsize())
print("Queue full check before inserting all elements: ",q.full())
print("Queue empty check before inserting all elements: ",q.empty())
# Do Insert Operation in queue

for i in range(1,6):
    value = input("Enter value to insert into queue")
    q.put(value)

#q.put('a')
#q.put(10)
#q.put('b')
#q.put(100)
#q.put("apple")

print("Queue size after inserting 5 elements: ", q.qsize())
print("Queue elements :")
print(list(q.queue))

print("Queue full check after inserting all elements: ",q.full())
print("Queue empty check after inserting all elements: ",q.empty())

# Do delete operation in queue
print("Deleted elements")
for i in range(1,6):
    print(q.get())
```

```python
class Queue:
    def __init__(self):
        self.items = []

    def enqueue(self, item):
        self.items.append(item)

    def dequeue(self):
        if not self.is_empty():
            return self.items.pop(0)
        else:
            raise IndexError("Dequeue from an empty queue")

    def peek(self):
        if not self.is_empty():
            return self.items[0]
        else:
            raise IndexError("Peek from an empty queue")

    def is_empty(self):
        return len(self.items) == 0

    def size(self):
        return len(self.items)

# Example usage
queue = Queue()
queue.enqueue(10)
queue.enqueue(20)
queue.enqueue(30)

print(f"Front element: {queue.peek()}")  # Output: 10
print(f"Queue size: {queue.size()}")     # Output: 3

print(f"Dequeued element: {queue.dequeue()}")  # Output: 10
print(f"Queue size after dequeue: {queue.size()}")  # Output: 2

print(f"Is queue empty? {queue.is_empty()}")  # Output: False
```
